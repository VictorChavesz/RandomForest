```python
#importação das bibliotecas
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#importação da base de dados
data = pd.read_csv("CreditCardFraud.csv", sep = ';')
data.head()
```

```python
#verificação se existem valores nulos
data.isnull().sum()
```

```python
#transformando os valores com , para .
data = data.replace(',','.', regex=True)
data = data.astype(float)
```

```python
#antes da validação da importância das features, vamos realizar um teste com as colunas originais
features = data.iloc[:,:-1].values
target = data.iloc[:,-1].values
```

```python
features
```

```python
target
```

```python
#separação dos dados em treino e teste
x_train,x_test,y_train,y_test = train_test_split(features, target, test_size = 0.2, random_state = 42)
```

```python
#criação e treino do modelo
RandomForest = RandomForestClassifier()
RandomForest.fit(x_train,y_train)
```

```python
#predição
forecast = RandomForest.predict(x_test)
```

```python
len(x_train)
```

```python
forecast
```

```python
y_test
```

```python
#verificação da taxa de acerto
accuracy_score(y_test,forecast)
print(f"A taxa de acerto do modelo é de {accuracy_score(y_test,forecast)*100}%")
```

```python
#mesmo com a taxa de acerto em 100%, vamos verificar a importancia das colunas
forest = ExtraTreesClassifier()
forest.fit(x_train,y_train)
```

```python
importance = forest.feature_importances_
importance
```

```python
importance_sorted = np.argsort(importance)[::-1]
```

```python
for i in importance_sorted:
    print(f"A importância da coluna {data.columns[i]} na posição {i} é de {importance[i]}")
```
